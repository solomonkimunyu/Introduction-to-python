A module is simply a file containing python code, it may contain functions, classes etc

## Creating a python module

```python
def add(a, b):
    return a + b
def sub(a, b):
    return a - b
def mul(a, b):
    return a * b
def divide(a, b):
    return a/b
```

### Calling a module in python

We use import keyword to incorporate the module into our program

```python
import mod1 as md
addition = md.add(10,15)
print(addition)
```

```python
from mod1 import * #Do this on pycharm
res =mul(10,15)
print(res)
```

```python
# A program for calculating factorial of a number
def fact(num):
    if num == 1:
        return num
    else:
        return num * fact(num - 1)
```

```python
def pal(x):
    x! = x[::-1]
    if x! == x:
        return 'palindrome'
    else:
        return 'not palindrome'
```
